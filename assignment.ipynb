# Clean Data

To be remove special characters (\n,\r)

```python
raw_data = [print_press,blog_post,about_us]
```

```python
raw_data
```

```python
data=[]
for i,v in enumerate(raw_data):
    process = v.replace("\n","")
    process = process.replace("\r","")
    process = " ".join(process.split())
    data.append(process)
```

```python
data
```

# RAG

## Setup model and embedding

```python
from llama_index.core import Settings
#from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.llms.ollama import Ollama
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

embed_name = "BAAI/bge-m3"
#llm_name = "scb10x/llama3.2-typhoon2-3b-instruct"
llm_name = "llama3.2"

embed_model = HuggingFaceEmbedding(model_name= embed_name)
llm = Ollama(model="llama3.2", request_timeout=3600.0)
#llm = HuggingFaceLLM(model_name=llm_name,tokenizer_name=llm_name,model_kwargs={"torch_dtype": "auto"})

Settings.llm = llm
Settings.embed_model = embed_model
```

## Tokenize Data

```python
from pythainlp.tokenize import word_tokenize
```

```python
token =[]
for v in data:
    temp=word_tokenize(v)
    process = " ".join(temp)
    token.append(process)
```

```python
token
```

## Vector Store

```python
from llama_index.core.node_parser import TokenTextSplitter
tts = TokenTextSplitter.from_defaults(chunk_overlap=100)
chunk=[]
for v in token:
    temp = tts.split_text(v)
    chunk.extend(temp)
```

```python
# split document in to chunk with 100 token overlap
chunk
```

```python
from llama_index.core import Document
# Convert python string to llama_index document
llama_index_doc = [Document(text=s) for s in chunk]
```

```python
# This cell is from first time data ingest
import chromadb
from llama_index.core import VectorStoreIndex,StorageContext
from llama_index.vector_stores.chroma import ChromaVectorStore

db = chromadb.PersistentClient(path="./vector_db")
chroma_collection = db.get_or_create_collection("qa_vector_store")
vector_store = ChromaVectorStore(chroma_collection=chroma_collection)
storage_context = StorageContext.from_defaults(vector_store=vector_store)

index = VectorStoreIndex.from_documents(llama_index_doc, storage_context=storage_context)
```

```python
# When you want to load existing data from disk
import chromadb
from llama_index.core import VectorStoreIndex,StorageContext
from llama_index.vector_stores.chroma import ChromaVectorStore

db = chromadb.PersistentClient(path="./vector_db")
chroma_collection = db.get_or_create_collection("qa_vector_store")
vector_store = ChromaVectorStore(chroma_collection=chroma_collection)
#storage_context = StorageContext.from_defaults(vector_store=vector_store)
index = VectorStoreIndex.from_vector_store(vector_store=vector_store)
```
